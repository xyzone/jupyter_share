```python
class BinaryTree(object):
    
    def __init__(self, rootObj):
        self.key = rootObj
        self.leftChild = None
        self.rightChild = None
        self.parent = None
        
    def insertLeft(self, newNode):
        t = BinaryTree(newNode) 
        if self.leftChild == None:
            self.leftChild = t
            t.parent = self
        else:                        
            t.parent = self
            self.leftChild.parent = t
            t.leftChild = self.leftChild 
            self.leftChild = t
                    
    def insertRight(self, newNode):
        t = BinaryTree(newNode) 
        if self.rightChild == None:
            self.rightChild = t 
            t.parent = self
        else:            
            t.parent = self
            self.rightChild.parent = t
            t.rightChild = self.rightChild
            self.rightChild = t
     
    def getRightChild(self):
        return self.rightChild
    
    def getLeftChild(self):
        return self.leftChild
    
    def is_leaf(self):
        return not (self.leftChild or self.rightChild)
    
    def setRootVal(self, obj):
        self.key = obj
        
    def getRootVal(self):
        return self.key
    
r = BinaryTree(10) 
r.insertLeft(7) 
lc = r.getLeftChild()
lc.insertLeft(4)
lc.insertRight(9)

llc = lc.getLeftChild()
llc.insertLeft(1)
llc.insertRight(6)
r.insertRight(14)

rc = r.getRightChild()
rc.insertLeft(12)
rc.insertRight(18)
print (r.getRightChild().getRightChild().getRootVal())


def parsePreOrder(tree_node):
    if tree_node.is_leaf():
        return [tree_node.getRootVal()]
    left_list = right_list = []
    if tree_node.getLeftChild():
        left_list = parsePreOrder(tree_node.getLeftChild())
    if tree_node.getRightChild():
        right_list = parsePreOrder(tree_node.getRightChild())
    return [tree_node.getRootVal()] + left_list + right_list   


def parseInOrder(tree_node):
    if tree_node.is_leaf():
        return [tree_node.getRootVal()]
    left_list = right_list = []
    if tree_node.getLeftChild():
        left_list = parseInOrder(tree_node.getLeftChild())
    if tree_node.getRightChild():
        right_list = parseInOrder(tree_node.getRightChild())
    return left_list + [tree_node.getRootVal()] + right_list   
         

def parsePostOrder(tree_node):
    if tree_node.is_leaf():
        return [tree_node.getRootVal()]
    left_list = right_list = []
    if tree_node.getLeftChild():
        left_list = parsePostOrder(tree_node.getLeftChild())
    if tree_node.getRightChild():
        right_list = parsePostOrder(tree_node.getRightChild())
    return left_list + right_list + [tree_node.getRootVal()]  

def checkBinarySearchTree(tree_node):
    node_value = tree_node.getRootVal()
    left_node = tree_node.getLeftChild()
    right_node = tree_node.getRightChild()
    if left_node:
        if left_node.getRootVal() > node_value:
            return False
        else:
            return checkBinarySearchTree(left_node)
    if right_node:
        if right_node.getRootVal() < node_value:
            return False
        else:
            return checkBinarySearchTree(right_node)
    return True

def levelOrderTree(tree_nodes, level):
    if not tree_nodes:
        return 'Completed Scan'
    print (level, [nd.getRootVal() for nd in tree_nodes])
    new_nodes = []
    for node in tree_nodes:
        if node.getLeftChild():
            new_nodes += [node.getLeftChild()] 
        if node.getRightChild():
            new_nodes += [node.getRightChild()]
    return levelOrderTree(new_nodes, level + 1)
    
def trimBinarySearchTree(tree_node, minLimit, maxLimit):
    node_val = tree_node.getRootVal()
    left_child = tree_node.getLeftChild()
    right_child = tree_node.getRightChild()
    
    if node_val < minLimit:
        # trim left sub tree and replace by right sub tree
        left_child = None
        if right_child:
            right_child.parent = tree_node.parent
            tree_node.parent.leftChild = right_child
        else:
            tree_node.parent.leftChild = None
        
    elif node_val > maxLimit:
        right_child = None
        if left_child:
            left_child.parent = tree_node.parent
            tree_node.parent.rightChild = left_child
        else:
            tree_node.parent.rightChild = None
            
    if left_child:
        trimBinarySearchTree(left_child, minLimit, maxLimit)
    if right_child:
        trimBinarySearchTree(right_child, minLimit, maxLimit)
    
    
print (parsePreOrder(r))
print (parseInOrder(r))
print (parsePostOrder(r))
print (checkBinarySearchTree(r))
print (levelOrderTree([r], 0))
print (trimBinarySearchTree(r, 5, 13))
print (levelOrderTree([r], 0))
```
